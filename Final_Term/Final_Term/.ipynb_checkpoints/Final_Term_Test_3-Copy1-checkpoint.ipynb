```python
from selenium import webdriver
from bs4 import BeautifulSoup
import bs4
import time
import re
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Chrome(r'C:\Users\skldd\anaconda3\envs\pytlesson\BD\chromedriver.exe')
```

```python
driver.get('https://www.mcdonalds.co.kr/kor/store/main.do')
time.sleep(5)
```

```python
element_class = driver.find_element_by_id('searchWord')
xpath = """//*[@id="searchForm"]/div/fieldset/div/button"""
element_search = driver.find_element_by_xpath(xpath).click()
```

```python
table_address = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr[1]/td[1]/dl/dd[1]")
table_address.text
```

```python
table_new_address = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr[1]/td[1]/dl/dd[2]")
table_new_address.text
```

```python
table_phone_number = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr[1]/td[2]")
table_phone_number.text.split('\n')
```

```python
table_office_hour = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr[1]/td[3]")
table_office_hour.text
```

```python
table_service = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr[1]/td[4]/div")
table_service.text
```

```python
total = []
for j in range(1,83,1):
    if (j%10) != 0:
        page_num = j%10
    else:
        page_num = 10
    xpath = "//*[@id='container']/div[1]/div[2]/div/div/div[2]/div/span/a[" + str(page_num) + "]"
    element_search = driver.find_element_by_xpath(xpath).click()
    time.sleep(2)
    table = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table")
    
    tr = table.find_elements_by_tag_name('tr')
    for i in range(1,len(tr),1):
        tmp = []
        table_name = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr["+ str(i)+"]/td[1]/dl/dt/strong/a")
        table_address = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr["+ str(i)+"]/td[1]/dl/dd[1]")
        table_new_address = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr["+ str(i)+"]/td[1]/dl/dd[2]")
        table_phone_number = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr["+ str(i)+"]/td[2]")
        table_office_hour = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr["+ str(i)+"]/td[3]")
        table_service = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/table/tbody/tr["+ str(i)+"]/td[4]/div")
    
    
        tmp.append(table_name.text)
        tmp.append(table_address.text)
        tmp.append(table_new_address.text)
        tmp.append(table_phone_number.text.split('\n'))
        tmp.append(table_office_hour.text)
        tmp.append(table_service.text.split('\n'))
    
        total.append(tmp)
    
    if (j%10) == 0:
        next_page = driver.find_element_by_xpath("//*[@id='container']/div[1]/div[2]/div/div/div[2]/div/a[3]").click()
        time.sleep(2)
total
```

```python
for i in range(len(total)):
    local = total[i][1].split(' ')
    if local[0] == '서울' or local[0] == '서울시':
        Seoul.append(total[i])
    elif local[0] == '인천':
        Incheon.append(total[i])
    elif local[0] == '울산':
        Ulsan.append(total[i])
    elif local[0] == '부산':
        Busan.append(total[i])
    elif local[0] == '대구':
        Daegu.append(total[i])
    elif local[0] == '대전':
        Daejeon.append(total[i])
    elif local[0] == '광주':
        Gwanju.append(total[i])
    elif local[0] == '제주특별자치도':
        Jeju.append(total[i])
    elif local[0] == '경기':
        Gyeonggi.append(total[i])
    elif local[0] == '강원':
        Gangwon.append(total[i])
    elif local[0] == '경남':
        GyeongNam.append(total[i])
    elif local[0] == '경북':
        GyeongBuk.append(total[i])
    elif local[0] == '전남':
        JeonNam.append(total[i])
    elif local[0] == '전북':
        JeonBuk.append(total[i])
    elif local[0] == '충남':
        ChungNam.append(total[i])
    elif local[0] == '충북':
        ChungBuk.append(total[i])
    else:
        print('No Local')
        print(total[i])
```
