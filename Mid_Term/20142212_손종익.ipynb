```python
def clean_text(text):
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>▶▽♡◀@\#$%&\\\=\(\'\'ⓒ(\n)(\t)]','', text)
    cleaned_text = re.sub(' ', '', cleaned_text)
    return cleaned_text
```

```python
def compare_title(title1, title2):
    cleaned_title1 = clean_text(title1)
    cleaned_title2 = clean_text(title2)
    title1 = list(cleaned_title1)
    title2 = list(cleaned_title2)
    count = 0
    for i in range(len(title1)):
        if title1[i] == title2[i]:
            count = count + 1
            if count >= 10:
                return 1
            
        elif title1[i] != title2[i]:
            return 0
```

```python
office_title_url = []
for link_find in range(len(urls)):
    for page in range(1,100,1):
        
        url_search = urls[link_find] + str(page)
        headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.87 Safari/537.36'}
        r = requests.get(url_search, headers = headers)
        bs = BeautifulSoup(r.text, 'html.parser')
        #time.sleep(1)
        
        find_title = bs.select('ul > li > dl > dt > a')
    
        title_url = []
        for z in range(len(find_title)):
            tmp = []
            if len(find_title[z].text.strip()) != 0:
                find_title_a = find_title[z].text.strip()
                find_url_a = find_title[z]['href']
                tmp.append(find_title_a)
                tmp.append(find_url_a)
                title_url.append(tmp)

        for y in range(len(title_url)):
            compare = compare_title(rank_list[link_find][1],title_url[y][0])
            if compare == 1:
                result = {}
                result['언론사'] = rank_list[link_find][0]
                result['제목'] = title_url[y][0]
                result['원문링크'] = title_url[y][1]
                office_title_url.append(result)
                break
                
office_title_url
```

```python
df = pd.DataFrame(office_title_url,columns = ['언론사','제목','원문링크'])
df.to_csv('news_office_title_url.csv', encoding='utf-8-sig')
```

```python
import os
import sys
import urllib.request
import datetime
import time
import json
import xmltodict
import math
import pandas as pd
from urllib.request import urlopen, Request
from urllib.parse import urlencode, quote_plus
from config_public import *
```

```python
def get_request_url(url):
    
    req = urllib.request.Request(url)
    
    try: 
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            print ("[%s] Url Request Success" % datetime.datetime.now())
            return response.read().decode('utf-8')
    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None
```

```python
def getUrl(start_date, end_date,code):
    
    end_point = "http://apis.data.go.kr/1140100/minAnalsInfoView2/minTopNKeyword2"
    parameters = "?serviceKey=" + access_key
    parameters += "&target=pttn,dfpt,saeol"
    parameters += "&dateFrom=" + str(start_date)
    parameters += "&dateTo=" + str(end_date)
    parameters += "&depCode=" + str(code)
    parameters += "&resultCount=10&searchCount=100"

    url = end_point + parameters
    
    retData = get_request_url(url)

    print(retData)
    if (retData == None):
        return None
    else:
        return json.loads(retData)
```

```python
def main():
    jsonResult = []
    nStartDate = 20190101
    nEndDate = 20190105
    code = 1620000
    jsonData = getUrl(nStartDate,nEndDate,code)
    jsonResult.append(jsonData)
    print(jsonData)
    df = pd.DataFrame(jsonResult)
    df.to_csv('민원.csv', encoding='utf-8-sig')
```
